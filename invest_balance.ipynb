# Invest Balance

```python
import matplotlib.pyplot as plt

monthly_invest = {
 # US
 "09SN": 300,
 "VAG2": 120,
 "VJZ9":  80,
 # Global
 "VGEU": 150,
 "0V18": 120,
 "SS02": 100,
 "EUNY":  50,
 # Specific
 "0V1Y":  80
}
amounts = list(monthly_invest.values())
total = sum(amounts)
explode = [0]*len(amounts)
explode[-1] = 0.1

colors=["dodgerblue","cornflowerblue","darkturquoise","yellow","firebrick","hotpink","mediumorchid","deepskyblue"]

plt.figure(0,figsize=(9,12))
plt.pie(monthly_invest.values(),labels=monthly_invest.keys(),autopct='%1.1f%%',
        pctdistance=.9, shadow = True, explode=explode, colors=colors)
plt.show()
```

```python
div_invest = [
 # US
 (30.0, 3.50, 0.06),
 (12.0, 3.00, 0.08),
 ( 8.0, 4.50, 0.07),
 # Global
 (15.0, 3.00, 0.10),
 (12.0, 2.75, 0.40),
 (10.0, 5.00, 0.08),
 ( 5.0, 7.00, 0.49),
 # Specific
 ( 8.0, 4.00, 0.12)
]

total_win,total_cost = (0,0)
for triplett in div_invest:
    total_win+=(triplett[0]/100)*triplett[1]
    total_cost+=(triplett[0]/100)*triplett[2]
print("Invest Dividend:",round(total_win,2))
print("Invest TER:",round(total_cost,2))
```
