# Invest Balance

```python
import matplotlib.pyplot as plt

monthly_invest = {
 # US
 "09SN": 420,
 "VAG2": 200,
 "VJZ9":  45,
 # Global
 "VGEU": 100,
 "0V18":  70,
 "SS02":  65,
 "EUNY":  20,
 # Specific
 "0V1Y":  80
}
amounts = list(monthly_invest.values())
total = sum(amounts)
explode = [0]*len(amounts)
explode[-1] = 0.1


plt.figure(0,figsize=(9,12))
plt.pie(monthly_invest.values(),labels=monthly_invest.keys(),autopct='%1.1f%%', pctdistance=.9, shadow = True, explode=explode)
plt.show()
```

```python
div_invest = [
 # US
 (42.0, 3.50, 0.06),
 (20.0, 3.00, 0.08),
 ( 4.5, 4.50, 0.07),
 # Global
 (10.0, 3.00, 0.10),
 ( 7.0, 2.75, 0.40),
 ( 6.5, 5.00, 0.08),
 ( 2.0, 7.00, 0.49),
 # Specific
 ( 8.0, 4.00, 0.12)
]

total_win,total_cost = (0,0)
for triplett in div_invest:
    total_win+=(triplett[0]/100)*triplett[1]
    total_cost+=(triplett[0]/100)*triplett[2]
print("Invest Dividend:",round(total_win,2))
print("Invest TER:",round(total_cost,2))
```
